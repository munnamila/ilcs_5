```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3, 3, 20)

y1 = x
y2 = x ** 2
y3 = x ** 3
y4 = x ** 4
```

```python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(6, 4))

axs[0, 0].plot(x, y1)
axs[0, 1].plot(x, y2)
axs[1, 0].plot(x, y3)
axs[1, 1].plot(x, y4)

plt.show()

```
